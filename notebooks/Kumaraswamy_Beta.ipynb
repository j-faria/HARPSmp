```python
import numpy as np
from kumaraswamy import kumaraswamy  # github.com/j-faria/Kumaraswamy
from scipy.stats import beta
```

```python
fig, ax = plt.subplots(1, 1)

x = np.linspace(0, 1, 1000)
# pdfs
ax.plot(x, kumaraswamy(0.867, 3.03).pdf(x), color='C0')
ax.plot(x, beta(0.867, 3.03).pdf(x), '--', color='C1', dashes=[6, 2])

axr = ax.twinx()
# cdfs
axr.plot(x, kumaraswamy(0.867, 3.03).cdf(x), color='C0')
axr.plot(x, beta(0.867, 3.03).cdf(x), '--', color='C1', dashes=[6, 2])

ax.legend(['Kumaraswamy', 'Beta'], loc='center right', frameon=False, fontsize=12)
ax.text(0.61, 2.1, r'$a=0.867, b=3.03$', ha='left', fontsize=12)


ax.set(xlabel='Eccentricity', ylabel='Probability density function');
axr.set(ylabel='Cumulative distribution function');
```
